```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 1: Convert dictionary to DataFrame
d = {"C": [12, 20, 34, 45, 55], "C++": [15, 25, 36, 50, 60], "Java": [18, 30, 40, 55, 70], "R": [20, 35, 45, 60, 80], "Python": [25, 40, 53, 75, 100]}
df = pd.DataFrame(d)

# Step 2: Convert text labels to uppercase
X1 = ["C", "C++", "Java", "Python", "R"]
Y1 = ["200", "400", "600", "800", "1000"]
X1_upper = [x.upper() for x in X1]
Y1_upper = [y.upper() for y in Y1]

# Step 3: Plot the data
plt.figure(figsize=(10, 6))

# Plot each line separately with custom markers
sns.lineplot(data=df["C"], marker="D", label="C", color="blue")        # Diamond
sns.lineplot(data=df["C++"], marker="s", label="C++", color="red")    # Square
sns.lineplot(data=df["Java"], marker="^", label="Java", color="green")  # Triangle
sns.lineplot(data=df["R"], marker="x", label="R", color="black")        # Cross
sns.lineplot(data=df["Python"], marker="*", label="Python", color="darkblue")  # Star

# Customizing the plot
plt.title('Performance Comparison')
plt.xlabel('File Size (MB)')
plt.ylabel('Time Taken (Sec)')
plt.xticks(ticks=np.arange(len(Y1)), labels=Y1_upper)
# plt.legend(title='Languages', loc='dow right')

plt.grid(True)
plt.show()
```
